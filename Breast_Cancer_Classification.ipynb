 In this project, we build a Breast Cancer Classification system using Logistic Regression with Python

Importing the Dependencies

```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data Collection and Processing

```python
# loading the data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
#print(breast_cancer_dataset)
```

```python
# Loading the data to a dataframe
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
# print the first 5 rows of the dataframe
data_frame.head()
```

```python
# adding the 'target' column to the dataframe
data_frame['label'] = breast_cancer_dataset.target
```

```python
# print the last 5 rows of the dataframe
data_frame.tail()
```

```python
# number of rows and columns in the dataset
data_frame.shape
```

```python
# getting some information about the dataset
data_frame.info()
```

```python
# checking for missing values
data_frame.isnull().sum()
```

```python
# Statistical measures about the data
data_frame.describe()
```

```python
# checking the distribution of target variables
data_frame['label'].value_counts()
```

  - 1 - Benign cases
  - 0 - Malignant cases

```python
data_frame.groupby('label').mean()
```

Splitting the features and target

```python
X = data_frame.drop(columns = 'label', axis = 1)
y = data_frame['label']
```

```python
print(X)
```

```python
print(y)
```

Splitting the data into Training and Testing data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_test.shape, X_train.shape)
```

Model Training
  - Logistic Regression

```python
model = LogisticRegression()
```

```python
# training the LogisticRegression model using training data
model.fit(X_train, y_train)
```

Model Evaluation
  - Accuracy score

```python
# Accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(y_train, X_train_prediction)
```

```python
print(training_data_accuracy)
```

```python
# Accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(y_test, X_test_prediction)
```

```python
print(test_data_accuracy)
```

Building a Predictive System

```python
input_data = [20.57,17.77,132.9,1326,0.08474,0.07864,0.0869,0.07017,0.1812,0.05667,0.5435,0.7339,3.398,74.08,0.005225,0.01308,0.0186,0.0134,0.01389,0.003532,24.99,23.41,158.8,1956,0.1238,0.1866,0.2416,0.186,0.275,0.08902]

# change the input data to a numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the numpy array as we are predicting for one datapoint
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

# Predict the label for the input data
prediction = model.predict(input_data_reshaped)

print(prediction)

if (prediction[0] == 0):
  print('The breast cancer is Malignant')
else:
  print('The breast cancer id Benign')
```
